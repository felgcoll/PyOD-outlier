```python
import io
import os
import pandas as pd
```

```python
#Funci√≥n para crear las columnas que van a enumerar el DF
def columns():
    col = []
    
    for i in range(1,9):
        col.append('x'+str(i))
        col.append('y'+str(i))
        col.append('z'+str(i))
    
    return col
```

```python
#Directorio a recorrer
path_csv = '/home/bryan/Escritorio/PyOD/Dataset/'

def Ficheros(path, formato):
    contenido = os.listdir(path)

    #Acumulador del nombre de las   imagenes a utilizar
    archivos = []

    #Se recorre el path para obtener los nombre y se los aggrega al acumulador, en este caso la lista
    for fichero in contenido:
        if os.path.isfile(os.path.join(path, fichero)) and fichero.endswith(formato):
            archivos.append(fichero)
            
    return archivos


#Verificar si obtiene los archivos
print(len(Ficheros(path_csv, '.csv')))
```

```python
sampleo = '/home/bryan/Escritorio/PyOD/Dataset/'
dataset = Ficheros(path_csv, '.csv')
```

```python
prueba_1 = []
prueba_2 = []
prueba_3 = []

for archivo in dataset:
    if archivo.startswith('1_1'):
        prueba_1.append(archivo)
        
    elif archivo.startswith('1_2'):
        prueba_1.append(archivo)
    
    elif archivo.startswith('1_3'):
        prueba_1.append(archivo)
    
    elif archivo.startswith('1_4'):
        prueba_1.append(archivo)
    
    elif archivo.startswith('1_5'):
        prueba_1.append(archivo)
        
    elif archivo.startswith('1_6'):
        prueba_1.append(archivo)
    
    elif archivo.startswith('1_7'):
        prueba_1.append(archivo)
        
    elif archivo.startswith('1_8'):
        prueba_2.append(archivo)
    
    elif archivo.startswith('1_9'):
        prueba_3.append(archivo)
        

    

prueba_1.sort()
prueba_2.sort()
prueba_3.sort()
```

```python
prueba_1
```

```python
def dataframe_chipeado(lista):
    df_HB={}
    prueba = columns()
    df_HB = pd.DataFrame(df_HB)
    for csv in range(len(lista)):
        df_t = pd.read_csv(sampleo+ lista[csv])
        a1 = df_t.iloc[:, 1:25]
        df_HB= pd.concat([df_HB, a1], ignore_index=True)
        
    df_HB.columns = prueba
    
    return df_HB

base_1 = dataframe_chipeado(prueba_1)

base_2 = dataframe_chipeado(prueba_2)

base_3 = dataframe_chipeado(prueba_3)
```

# Graficos

```python
import seaborn as sns
import matplotlib.pyplot as plt

columnas = 24


for columna in range(columnas):
        ax = sns.histplot(data=base_1.iloc[:,columna], kde=True, color = 'green')
        ax = sns.histplot(data=base_1x.iloc[:,columna], kde=True, color = 'blue')
        ax = sns.histplot(data=base_2x.iloc[:,columna], kde=True, color = 'yellow')
        ax = sns.histplot(data=base_3x.iloc[:,columna], kde=True, color = 'red')
        
        print('Sensor: '+str(columna+1))
        plt.grid(True)
        plt.show()
```

```python
import numpy as np
```

```python
df_new = pd.concat([base_1, base_1x, base_2x, base_3x], ignore_index=True)
```

```python
df1_corr = df_new.corr()

mask = np.triu(np.ones_like(df1_corr, dtype=np.bool))

mask = mask[1:, :-1]
corr = df1_corr.iloc[1:,:-1].copy()

cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)

fig, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(corr, mask=mask, annot=True, cmap=cmap, vmin=-1, vmax=1, square=True)
#plt.savefig('Hb.png')
```

```python
hb = df_new.transpose()

from scipy.cluster.hierarchy import dendrogram, linkage
#sample1 = hb.sample(n = 10000)
#sample1.head()

Z = linkage(hb, 'ward')

plt.figure(figsize=(15, 5))
dn = dendrogram(Z, labels = hb.index ,orientation='right')
#plt.savefig('HB.jpg')
```
